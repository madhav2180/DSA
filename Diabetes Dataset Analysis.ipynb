```python
#Univariate Analysis
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv('diabetes_model.csv')
plt.hist(df[df['Outcome']==0]['Age'], bins = 5)
plt.title('Distribution of Age for Women who do not have Diabetes')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(df['Age'])

plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()
```

```python
#Bivariate Analysis
```

```python
import seaborn as sns
sns.scatterplot(x='Glucose',y='Insulin',data=df)
plt.show()
```

```python
corr_matrix = df.iloc[:,0:8].corr()

corr_matrix
```

```python
a = df['Age'].values
b = df['BloodPressure'].values
plt.hist(a,bins=10,color='green')
plt.hist(b,bins=10,color='red',alpha=0.3)
plt.xlabel('insulin')
plt.ylabel('glucose')
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
#Multivariate Analysis
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection="3d")
x=df['Pregnancies'].values
y=df['Age'].values
z=df['BloodPressure'].values
ax.plot3D(x, y, z, 'red')
```

```python
sns.pairplot(data=df,vars=['Glucose', 'SkinThickness', 'DiabetesPedigreeFunction'], hue = 'Outcome')
plt.show()
```
